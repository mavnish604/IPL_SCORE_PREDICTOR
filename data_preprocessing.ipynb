```python
import pandas as pd
import numpy as np
```

```python
in_df = pd.read_csv("ipl_ball_by_ball.csv")
```

```python
in_df
```

```python
in_df.describe()
```

```python
in_df['wicket_fallen'] = in_df['player_dismissed'].notna().astype(int)
in_df['wickets_lost'] = in_df.groupby(['match_id', 'innings'])['wicket_fallen'].cumsum()
in_df['wickets_remaining'] = 10 - in_df['wickets_lost']
cols_to_sum = ["runs_off_bat", "wides", "noballs", "byes", "legbyes", "penalty"]
in_df[cols_to_sum] = in_df[cols_to_sum].apply(pd.to_numeric, errors='coerce')
in_df["runs_scored"] = in_df[cols_to_sum].fillna(0).sum(axis=1)
```

```python
in_df
```

```python
df = in_df[['match_id', 'venue', 'innings', 'ball',
            'batting_team', 'bowling_team', 'wickets_remaining','runs_scored']]
```
