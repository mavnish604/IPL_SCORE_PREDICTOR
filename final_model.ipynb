```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("final_scores.csv")
```

```python
df
```

```python
df["score"]=0
df["runs_scored"] = df["runs_scored"].astype(float)
final_scores = df.groupby(["match_id", "batting_team"])["runs_scored"].sum().reset_index()
final_scores.rename(columns={"runs_scored": "final_score"}, inplace=True)
df = df.merge(final_scores, on=["match_id", "batting_team"], how="left")
df
```

```python
def update_running_score(df):
    df["score"] = 0 
    for (match, innings), group in df.groupby(["match_id", "innings"]):
        total_score = 0 
        for idx in group.index:
            total_score += df.at[idx, "runs_scored"]
            df.at[idx, "score"] = total_score  

    return df
```

```python
df = update_running_score(df)
```

```python
df
```

```python
df["runs_last_5_overs"] = (
    df.groupby(["match_id", "innings"])["runs_scored"]
    .rolling(window=30, min_periods=1)
    .sum()
    .reset_index(level=[0, 1], drop=True)
)
df = df[df["ball"] >= 5.0].reset_index(drop=True)
```

```python
df
```

```python

df["balls_bowled"] = (df["ball"].astype(int) * 6) + ((df["ball"] % 1) * 10)

# Calculate remaining balls
df["balls_remaining"] = 120 - df["balls_bowled"]
```

```python
df=df.drop(columns=["balls_bowled","ball","match_id"])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df=df.drop(columns=["runs_scored"])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df['overs_played'] = (120 - df['balls_remaining']) / 6
df['crr'] = df['score'] / df['overs_played']
df=df.drop(columns=["overs_played"])
df
```

```python
df["balls_remaining"] = df["balls_remaining"].apply(lambda x: 0 if np.abs(x) < 1e-10 else int(x))
final_df = df.sample(df.shape[0])
```

```python
from sklearn.model_selection import train_test_split
X=final_df.drop(columns=['final_score'])
y= final_df['final_score']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
cltrf = ColumnTransformer([
    ('cltrf',OneHotEncoder(sparse_output=False,drop='first'),["batting_team","bowling_team","venue"])
]
,remainder="passthrough")
```

```python
model2 = Pipeline(steps=[
    ('step1',cltrf),
    ('step2',StandardScaler()),
    ("step3",XGBRegressor(n_estimators=1500,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
print("r2 score: ",r2_score(y_test,y_pred))
print("MEAN ABSOLUTE ERROR: ",mean_absolute_error(y_test,y_pred))
```
